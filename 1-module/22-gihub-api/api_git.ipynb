```python
def traducir_hora(hora):
    try:
        hora = float(hora)
    except:
        hora = re.findall(r"\d.*\d",hora)[0]
        hora = datetime.strptime(hora, "%d %b %Y %X").utctimetuple()
        hora = calendar.timegm(hora)
    return time.ctime(hora)
```

```python
traducir_hora(reset)
```

```python
traducir_hora(respuesta.headers['Date'])
```
