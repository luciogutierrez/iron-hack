```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(88)
df = pd.DataFrame({'A':np.linspace(1,10,10)})
df
```

```python
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))], axis=1)
df
```

```python
df.iloc[3, 3] = np.nan
df.iloc[0, 2] = np.nan
df
```

```python
# highlight max
df.style.highlight_max()
```

```python
# highlight min
df.style.highlight_min()
```

```python
# highlight nan
df.style.highlight_null(null_color='green')
```

```python
# highlight out of scope
df.style.set_na_rep('OutofScope').highlight_null(null_color='orange')
```

```python
# creating a heatmap within dataframe
df.style.background_gradient()
```

```python
# style as a table
df.style.set_properties(**{'border':'1.3px solid green', 'color':'magenta'})
```

```python
# creating a bar chars
df.style.bar()
```

```python
# float value precision
df.style.set_precision(2)
```

```python
df.style.set_caption('This is analytics vidhya blog').set_precision(2).background_gradient()
```

```python
# hiding index or column
df.style.hide_index()
```

```python
# styling % and decimals
df.style.format('{:.2%}', na_rep='').hide_index().background_gradient()
```

```python
# creating a personal highlight
def highlight_mean_greater(s):
    # higlight yellow is value greater than mean else read.
    is_max = s > s.mean()
    return ['background-color:yellow' if i else 'background-color:red' for i in is_max]

df.style.apply(highlight_mean_greater)
```

```python
# highlight values
def color_negative_red(val):
    color = 'red' if val < 0 else 'white'
    return f'color: {color}'

df.style.applymap(color_negative_red)
```

```python
# formating a table
styles = [
    dict(selector="tr:hover",
                props=[("background", "#f4f4f4")]),
    dict(selector="th", props=[("color", "#fff"),
                               ("border", "1px solid #eee"),
                               ("padding", "12px 35px"),
                               ("border-collapse", "collapse"),
                               ("background", "#00cccc"),
                               ("text-transform", "uppercase"),
                               ("font-size", "18px")
                               ]),
    dict(selector="td", props=[("color", "#999"),
                               ("border", "1px solid #eee"),
                               ("padding", "12px 35px"),
                               ("border-collapse", "collapse"),
                               ("font-size", "15px")
                               ]),
    dict(selector="table", props=[
                                    ("font-family" , 'Arial'),
                                    ("margin" , "25px auto"),
                                    ("border-collapse" , "collapse"),
                                    ("border" , "1px solid #eee"),
                                    ("border-bottom" , "2px solid #00cccc"),                                    
                                      ]),
    dict(selector="caption", props=[("caption-side", "bottom")])
]
```

```python
df.style.set_table_styles(styles).set_caption("Image by Author (Made in Pandas)").highlight_max().highlight_null(null_color='red')
```

```python
# Exporting to excel
df.style.set_precision(2).background_gradient().hide_index().to_excel('styled.xlsx', engine='openpyxl')
```
