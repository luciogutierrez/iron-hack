```python
def mediana_python(objeto):
  n = len(objeto)
  if n%2 != 0:
    mediana = sorted(objeto)[int((n-1)*0.5)]
  else:
    ob_ord = sorted(objeto)
    ind = int(n*0.5)
    mediana = (ob_ord[ind] + ob_ord[ind-1])/2
  return mediana
```
