## Conceptos previos

**Experimentos**


*   **Observaciones**: Se refiere a únicamente recolertar información ya sea del pasado o presente, sin intervención humana. \\
Por  ejemplo más detallado sería el registro de estatura de un adolescente, el simple dato de cuanto mide es una observación (170 cm). No necesariamente debe ser numérico, por ejemplo si sus ojos son verdes, eso también es una observación.


*   **Experimentos**: Son escenarios manipulados por humanos donde las variables significativas son controladas buscando obtener algún resultado especifico. \\
Por ejemplo, el dividir a un grupo de personas en 2 y a cada uno proporcionarle agua con azucar y agua con stevia. Con esto podríamos obtener la conclusión si la STEVIA es un sustitudo del azucar o no. 
Aquí involucra intervención humana.


¿Que es un estudio obervacional? \\
Involucara la recolección de varias observaciones sin que el humano intervenga.

---

**Población y Muestreo**



*   Población: Es un conjunto de datos que comparten características similares. 
Por ejemplo, si queremos ver cuantss tienen auto en México, nuestra población serían todos los habitantes de México \\
*   Muestra: Es un subconjunto de de nuestra población total la cual es usada para representar a toda la población en nuestro análisis estadístico.----

Sin embargo, con cualquier tipo de muestreo puedenocurrir lo que se conoce como **bias**.



---

**BIAS** \\

Bias: Se da cuando se hace un muestre de una pblación y pude generar los siguietnes escenarios:
* El muestreo que se elige puede no ser representativo del total de la población (ya sea accidentalmente o deliveradamente)

Un ejemplo de esto podría ser las encuestas en los super: Consideran que las personas que responden sus encuestas con usuarias del producto pero estan omitiendo que a lo mejor ni conocen el producto o aquellas personas que por x o y razón no pudieron contestar.

Hay varias maneras de reducir el BIAS:



*   Experimentos a ciegas: Por ejemplo, cuando se trata a un tipo de paciente con cierto medicamento y se le dice que esta recibiendo (no a ciegas) contra otro tipo de paciente que no sabe que se le esta administrando (a ciegas). 
Efecto placebo 
*   Exprimento "Double Blind": Mismo ejemplo que el anterior, solo que esta vez tanto el paciente como el que esta aplicando la prueba no conoce a quién si se le esta asignando un tratamiendo con efecto o no.

---


También podemos obtener "**contaminación**" en nuestras pruebas las cuales nos van a llevar a tener confusión en nuestras muestras. 
Siguiendo el ejemplo del efecto placebo, puede que nosotros esperemos un resultado determinado en aquellos pacientes que sí se diagnosticaron con el medicamente y cierto resultado en aquellos que no, sin embargo al llegar a tener resultados de manera similar entre ambos grupos, se nos dificulta el diferenciar entre aquellos que sí fueron medicados con placebo y aquellos que no.
El problema recae en que perdemos hábilidad en distinguir de manera correcta entre los efectos del placebo (los cuales llevan muchas otras variables) y los efectos del medicamento.
Si esto llegar a ser muy drastico, nuestro experimento puede anularse.

---

Para ello podemos usar la técnica de **Blocking**, la cual consiste en reducir el impacto de variables que nos pueden causar confusión separandolas en grupos.  

Por ejemplo, si de nuestro experimento con placebos y medicamentos separamos por la muestra por personas que ya tomaban algún tipo de medicamento anteriormente y en otro a aquellas que no, podríamos estudiar a cada grupo por aparte y tener conclusiones más sencillas de validar.

Para esta separación de nuestra muestra,  existen diferentes tecnicas para elegir nuestras muestras de población como:
1.   Muestreo aleatorio: totalmente aleatorio de nuestra muestra.
2.   Muestreo sistemático: elegimos de nuestra población miembros de 2 en 2 por ejemplo hasta un número determinado
3.   Muestreo estratificado: Dividimos nuestra población según alguna característica y de estos subgrupos vamos eligiendo nuestras muestra
4.   Muestreo por clústers: Seleccionamos grupos de maneare aleatoria y los categorizamos.









# Prueba de hipótesis
Una vez que ya aplicamos nuestras técnicas para seleccionar nuestra muestra de la población, podemos iniciar a probar nuestras hipótesis.

En **pruebas de hipótesis** tenemos una teoría acerca de algo y buscamos validar si nuestra teoría es correcta o no. 

**Tipos de hipótesis** \\
En cualquier modelo estadístico demos considerar siempre una hipótesis nula la cual se denomina $H_0$ y es la que niega nuestra nuestra afirmación principal. También podemos tener hipótesis alternativas denominadas como $H_1$ o $H_A$ las cuales son posibles resultados que esperamos obtener o corroborar.
Siempre consideraremos $H_0$ como verdadera hasta tener evidencia suficiente para negarla, entonces si aceptamos nuestras hipotesis alternativas.

---

**P - Value** \\
Debemos considerar nuestro $p-value$ el cual nos da la probabilidad de encontrar los datos de esta manera si nuestra hipótesis nula es verdadera.
Tipicamente nuestro nivel de significancia lo proponemos en 1%, 5% o 10%. Si el $p-value$ es muy bajo (cae en esos niveles), podríamos asegurar que la hipótesis nula es falsa y, por ende, nuestraa hipótesis alternativas (alguna o todas) son verdaderas.

Existen dos **tipos de Pruebas de Hipótesis**:
* Hipótesis direccional: únicamente cuando el valor de la prueba es muy alto, rechazamos nuestra hipótesis nula
* Hipótesis no direccional: Sucede cuando rechazamos la hipótesis nula si el valor es muy bajo o alto

**Intervalo de confianza** 

Los intervalos de confianza nos definen el rango que creemos en el que se encontraran nuestros valores de la muestra. Normalmente usamos un intervalo de confianza del 95% pero se pude calcular para cualquier porcentaje menor a 100%.

Se calcula así:

Mejor estimación +- margen de error \\
margen de error = z* Error estándar estimado

**Tipos de error**

En prueba de hipótesis puede ocurrir los siguientes tipos de error:
* Tipo I: La hipótesis nula es verdadera, pero la rechazamos (falso positivo). Esta probabilidad la conocemos como $\alpha$
* Tipo II: La hipótesis nula es falsa pero no la rechazamos (falso negativo). Se le conoce como $\beta$

¿Cual debemos de intentar hacer más pequeña?

Depende, por ejemplo:

* Si tenemos un falso positivo en un juicio, la hipótesis nula es que el defendido es inocente pero dado que es falso positivo, el inocente va a prisión.
* Si tenemos un falso negativo en una vacuna, consideramos nuestra hipótesis nula como que un paciente ya no está infectado. Esto daría a que dijéramos que el paciente se encuentra bien, pero lo consideramos como infectado (falso positivo) contra que el paciente se encuentre infectado y creamos que está bien (falso negativo). Cuando tenemos estos casos, queremos un $\beta$ bajo

```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```
plt.figure(figsize=(12,10))


plt.fill_between(x=np.arange(-4,-2,0.01), 
                 y1= stats.norm.pdf(np.arange(-4,-2,0.01)) ,
                 facecolor='red',
                 alpha=0.35)

plt.fill_between(x=np.arange(-2,2,0.01), 
                 y1= stats.norm.pdf(np.arange(-2,2,0.01)) ,
                 facecolor='grey',
                 alpha=0.35)

plt.fill_between(x=np.arange(2,4,0.01), 
                 y1= stats.norm.pdf(np.arange(2,4,0.01)) ,
                 facecolor='red',
                 alpha=0.5)

plt.fill_between(x=np.arange(-4,-2,0.01), 
                 y1= stats.norm.pdf(np.arange(-4,-2,0.01),loc=3, scale=2) ,
                 facecolor='grey',
                 alpha=0.35)

plt.fill_between(x=np.arange(-2,2,0.01), 
                 y1= stats.norm.pdf(np.arange(-2,2,0.01),loc=3, scale=2) ,
                 facecolor='blue',
                 alpha=0.35)

plt.fill_between(x=np.arange(2,10,0.01), 
                 y1= stats.norm.pdf(np.arange(2,10,0.01),loc=3, scale=2),
                 facecolor='grey',
                 alpha=0.35)

plt.text(x=-0.8, y=0.15, s= "Null Hypothesis")
plt.text(x=2.5, y=0.13, s= "Alternative")
plt.text(x=2.1, y=0.01, s= "Type 1 Error")
plt.text(x=-3.2, y=0.01, s= "Type 1 Error")
plt.text(x=0, y=0.02, s= "Type 2 Error");
```
