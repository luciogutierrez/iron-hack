# Calculando probabilidades/posibilidades (odds)

Las probabilidades de un evento son la relación entre la probabilidad de que suceda el evento y la probabilidad de que no suceda

$odds = \frac{P(A)}{1-P(A)}$

En el ejemplo de obtener un As lo podemos ver de la siguiente manera:
Sabemos que existen 4 cartas que son As y el mazo tiene 52 cartas por lo tanto tenemos 48 cartas que no son As.

y tendríamos 4 exitos contra 48 fracasos, si quisieramos obtener un As
y lo podríamos ver de la siguiente manera:

$ Odds (As) = 4:48 $

```
def odds(prob_eve):
  odds_perc = (prob_eve/(1-prob_eve))*100
  return round(odds_perc,1)
```

```
probabilidad = prob(4)
posibilildad = odds(4/52)
posibilildad_no_as = odds(48/52)

# Probabilidades los vemos como que oportunidad tengo de obtener determinado resultado entre todo mi espacio muestral
# Los ODDS los vemos como éxitos vs fracasos

print(f"Probabilidad de obtener un As:{probabilidad}%")
print(f"Posibilidad de obtener un As como exito:{posibilildad}%")
print(f"Posibilidad de no obtener un As como exito:{posibilildad_no_as}%")
```

```
corazones = 13
prob_corazon = prob(corazones)
print(f"Probabilidad de obtener un Corazon:{prob_corazon}%")
posib_corazon = odds(13/52)
print(f"Posibilidad de obtener un Corazon:{posib_corazon}%")
```

```
monos = 12
prob_monos = prob(monos)
print(f"Probabilidad de obtener un mono:{prob_monos}%")
posib_monos = odds(12/52)
print(f"Posibilidad de obtener un mono:{posib_monos}%")
# 12:40
# 6:20
# 3:10
```

### Permutaciones

las permutaciones representan el número de diferentes formas posibles de organizar un conjunto de elementos

$P(n, k),\ nPk=\frac{n!}{(n-k)!}$

Las permutaciones
1. Se toman todos los elementos de un conjunto
2. No se repiten los elementos del conjuntos
3. El orden importa ({A,B} es diferente a {B,A})

```
from scipy.special import perm
#De nuestros 4 aces, solo queremos tomar 2 de ellos. 
# n = 4, k = 2
perm(4,2)
perm()
```

### Combinaciones
En las combinaciones el orden no importa, ya que las combinaciones se refieren a las diferentes formas en las que se puede obtener un subconjunto de tamaño especifico dado un conjunto

$C(n, k), \ nCk= \frac{n!}{k!(n-k)!}$

```
from scipy.special import comb
comb(4,2)
```

Basicos de texas holdem:
1. Se juega con 52 cartas
2. Cada jugador recibe 2 cartas
3. Cada jugador intentara hacer el mejor juego con las cartas que recibe (su mano) y las 5 cartas que estarán saliendo a lo largo del juego.
4. Las cartas se van repartiendo en 4 rondas

Las rondas funcionan de la siguiente manera:
1. Pre-flop: A cada jugador se le dan dos cartas
2. Flop: Tres cartas comunitarias (al centro de la mesa) son repartidas
3. Turn: Una carta comunitaria es repartida
4. River: La última carta comunitaria es repartida

Cual es la probabilidad de obtener un Flush?

La forma más sencilla de calcular esto es calcular la probabilidad de que en el turn no obtengamos un diamante y en el river tampoco y restarle uno a este resultado

1. Saliera un diamante en turn y diamante en river
2. No saliera diamante en turn y si diamante en river
3. No saliera diamante en turn y no saliera diamante en river
4. Saliera diamante en turn y no saliera diamante en ricer

```
cartas = 52
mano = 2
cartas_centro = 3
cartas = cartas - (mano + cartas_centro)

# Diamantes para flush
diamantes = 13
diamantes_repartidos = 4
diamantes_faltantes = diamantes - diamantes_repartidos
# Cartas
cartas_no_diamantes = cartas - diamantes_faltantes

no_diamante_en_turn =  cartas_no_diamantes / cartas

#Calcular la prob de que no me salga un diamante en el river
cartas = cartas - 1
cartas_no_diamantes = cartas_no_diamantes - 1
no_diamante_en_river = cartas_no_diamantes/cartas

no_flush_river = no_diamante_en_turn * no_diamante_en_river

prob_river_flush = 1 - no_flush_river

print(f"Probabilidad de no flush en river : {no_flush_river}")
print(f"Probabilidad de flush en river : {prob_river_flush}")
```
