# Standard Uniform Distribution

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)

fig, ax = plt.subplots(1,1)
ax.plot(x, uniform.pdf(x));
```

### the uniform distributioncan also be discrete. Consider example of rolling a die. probabilito of each event happening is the same (1/6)

## Uniform distribution between point (a,b)

```python
a = 1
b = 3
x = np.linspace(a,b,100)
y = uniform.pdf(x, a, b)

fig, ax = plt.subplots(1,1)
ax.plot(x, y);
```

# Normal Distribution

## Normal Distribution is also known as the bell shape curve.

### Some important properties of normal distribution are:
### 1- It is symmetrical about the mean
### 2- Total area under the curve is 1
### 3- Mean and standard diviation can take on any real number
### 4- When =0 and =1, it is called a Standard Normal Deviation
### 'x' is referred to a 'z'/standardized x
### In a standard normal table has area to the left of z ie CDF value
### The area under the standar normal curve to the lesft of z-score gives the P(z < z-score)


```python
import math
from scipy.stats import norm
```

```python
normal = norm(0, math.sqrt(9))
print(normal.pdf(4))
print(normal.cdf(2))
print(normal.rvs())
x = np.linspace(-3,3,1000)
y = norm.pdf(x)

fig, ax = plt.subplots(1,1)
ax.plot(x,y)
plt.show();
```

# Exponential Distribution

```python
from scipy.stats import expon
```

```python
exp = expon(1)
print(exp.pdf(4))
print(exp.cdf(2))
print(exp.rvs())
x = np.linspace(0,10,100)
y = expon.pdf(x)
fig, ax = plt.subplots(1,1)
ax.plot(x, y)
```
