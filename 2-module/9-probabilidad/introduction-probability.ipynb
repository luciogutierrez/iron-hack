```python
# probabilidad de que al tirar un dado el resultado sea par esta dado por:
def even(x):
    return(x % 2 == 0)

sample_space = [1,2,3,4,5,6]
die_sides = len(sample_space)
even_roll = len([x for x in sample_space if even(x)])
even_probability = even_roll / die_sides
print(even_probability)
```

# Distribuciones de probabilidad discretas

## Variable aleatoria de Bernoulli

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p = 0.8
X = bernoulli.rvs(p, size=100)
plt.hist(X, align='mid')
```

## Variable Binomial random

```python
%matplotlib inline
from scipy import stats
from scipy.stats import binom
import numpy as np
```

```python
n = 10
p = 0.4
binomial = binom(n, p)
print(binomial.pmf(3)) # Probability Mass Fuction
print(binomial.cdf(3)) # Cumulative Distribution Function
print(binomial.mean())
print(binomial.var())
print(binomial.std())
print(binomial.rvs()) # Random Variables
x = np.arange(0, 10)
fig, ax = plt.subplots(1,1)
ax.plot(x, binom.pmf(x, n, p), 'bo')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5);

```

## Variable aleatoria de Poisson

```python
%matplotlib inline
from scipy import stats
from scipy.stats import poisson
```

```python
param = 2
po = stats.poisson(param)
print(po.pmf(param))
print(po.rvs(10))
x = np.arange(0,6)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, poisson.pmf(x, param), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, param), colors='b', lw=5, alpha=0.5);
```
