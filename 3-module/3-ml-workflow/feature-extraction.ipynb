```python
import pandas as pd
import numpy as np
```

```python
census = pd.read_csv('../../datasets/census.csv')
census.head()
```

```python
census.dtypes
```

```python
census.Income.hist()
```

```python
census['IncomeRange'] = pd.cut(census.Income, 3)
census.IncomeRange.value_counts()
```

```python
census['IncomePercentile'] = pd.qcut(census.Income, [0, .3333, .6666, 1])
census.IncomePercentile.value_counts()
```

```python
census.corr()
```

## Como se puede ver hombres y mujeres tienen una correlaciÃ³nd e .999527 lo que significa que podemos eliminar una de ellas sin afectar el modelo

```python
census['HighUnemploymentLowIncome'] = np.where((census.Unemployment > 0.07) & (census.IncomePercentile.cat.codes == 0), 1, 0)
census.HighUnemploymentLowIncome.value_counts()
```
