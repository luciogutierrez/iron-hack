```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def true_fxn(x):
    # sine wave function
    return np.sin(2 * np.pi * x)

np.random.seed(0)
```

```python
# generamos puntos aleatorios centrados alrededor de la función de verdad con un bit 
# de ruido aleatorio añadido.
num_samples = 30
noise_strength = 0.1

X = np.sort(np.random.rand(num_samples))
y = true_fxn(X) + np.random.randn(num_samples) * noise_strength
```

```python
plt.figure(figsize=(10,6))
x_range = np.linspace(0, 1, 100)
plt.plot(x_range, true_fxn(x_range), label='True function')
plt.scatter(X, y, label='Samples', edgecolor='r')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(PolynomialFeatures(degree=40), LinearRegression())
model.fit(X[:, np.newaxis], y)
```

```python
plt.figure(figsize=(10,6))
X_range = np.linspace(0, 1, 100)
plt.plot(X_range, true_fxn(X_range), label='True function')
plt.plot(X_range, model.predict(X_range[:, np.newaxis]), label='model')
plt.scatter(X, y, label='Sample', edgecolor='r')
plt.legend(loc='best')
plt.ylim(-2, 2)
plt.show()
```

```python
model.score(X[:, np.newaxis], y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.reshape(30,1), y, test_size=0.3)
```

```python
model.fit(X_train, y_train)
print('rendimiento del modelo con el set de entrenamiento', model.score(X_train, y_train))
print('rendimiento del modelo con el set de prueba', model.score(X_test, y_test))
```

```python
plt.figure(figsize=(10, 6))
X_range = np.linspace(0, 1, 100)
plt.plot(X_range, true_fxn(X_range), label="True Function")
plt.plot(X_range, model.predict(X_range[:, np.newaxis]), label="model")
plt.scatter(X_train, y_train, label="Training Samples", color='g')
plt.scatter(X_test, y_test, label="Testing Samples", color='r', marker="+")
plt.legend(loc="best")
plt.ylim((-2, 2))
plt.show()
```
