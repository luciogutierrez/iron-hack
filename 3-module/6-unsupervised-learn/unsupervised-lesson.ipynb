# DBSCAN

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
# Cargamos 2 medias lunas con un ruido de 005 para que las formas se distorcionen un poco
# y ver si lo detecta el modelo
moons = make_moons(n_samples=100, shuffle=True, noise=0.05)
plt.scatter(moons[0][:, 0], moons[0][:, 1]);
```

```python
# aplicamos DBSCAN para ver como los identifica
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.3).fit(moons[0])
dbscan.labels_
```

```python
plt.scatter(moons[0][:, 0], moons[0][:, 1], c=dbscan.labels_);
```

# GMM

```python
# Generamos los datos utilizando la funci√≥n make_classification
from sklearn.datasets import make_classification
from sklearn import mixture
```

```python
X = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1)
plt.scatter(X[0][:, 0], X[0][:,1], marker='*', s=25, edgecolor='k');
```

```python
# se alcanza a distiguir 2 grupos de datos
# GMM puede dectear esto de manera probabilistica
gmm = mixture.GaussianMixture(n_components=2, covariance_type ='full').fit(X[0])
plt.scatter(X[0][:, 0], X[0][:, 1], marker='^', c=gmm.predict(X[0]), s=25, edgecolor='k');
```
