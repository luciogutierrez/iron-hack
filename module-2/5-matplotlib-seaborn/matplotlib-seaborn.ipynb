# Matplotlib

```python
# garantiza que las visualizacioines se representen en el propio block de notas.
%matplotlib inline  
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)
plt.plot(x1, x1, label='linear')
plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Linear vs Non Linear plots')
plt.legend()
plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(1,2)

ax1.plot(x1,x2)
ax1.set_title('LINEAR PLOT')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

x = np.random.randn(10)
y = np.random.randn(10)
ax2.plot(x, y)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

## Correlacin Positiva - Cuando una variable aumenta con el aumento de otra y viceversa.
## Correlacin Negativa - Cuando una variable disminuye con el aumento de otra y viceversa.

```python
x = np.random.randn(10)
y = np.random.randn(10)
plt.scatter(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
```

# Grficos Circulares

```python
labels = ['Vegetariasn','Non-Vegetariasn','vegans']
sizes = [40,45,15]
```

```python
explode = (0,0.1,0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()

```
