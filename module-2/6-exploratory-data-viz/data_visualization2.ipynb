```python
%matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import seaborn as sns
pd.set_option('display.max_colwidth',100)
```

```python
diccionario = pd.read_csv('../../datasets/Diccionario de datos ILE.csv')
diccionario.head()
```

```python
jn = pd.read_json('../../datasets/interrupcion-legal-del-embarazo-20201204.json')
jn.head()
```

```python
df = json_normalize(jn.fields)
df.head()
```

```python
df.to_csv('../../datasets/int_legal_embarazo.csv')
```

```python
df['nile'] = df['nile'].replace('NA',0)
```

```python
# df = df.astype({'nile':'int32'})
df['nile'] = df['nile'].astype('int32')
```

```python
sns.set(context='paper', style='darkgrid', rc={'figure.figsize': (20, 10)})
ax = sns.boxplot(x=df['ocupacion'], y=df['edad'], hue=df['c_dolor'] != 'NA')
ax.set_title('Ocupación y Edad')
ax.set(xlabel='ocupación', ylabel="edad")
ax.set_xticklabels(ax.get_xticklabels(), rotation=65);
```

```python
sns.set(context='talk', style='darkgrid', rc={'figure.figsize':(20,10)})
ax = sns.boxplot(x=df['anticonceptivo'], y=df['edad'])
ax.set_title('Anticonceptivo y Edad')
ax.set(xlabel='Anticonceptivo', ylabel='Edad')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
sns.set(context='talk', style='darkgrid', rc={'figure.figsize':(20,10)})
ax = sns.boxplot(x=df['edad'], y=df['nile'].sort_values())
ax.set_title('Edad y No. de interrupciones')
ax.set(xlabel='Edad', ylabel='Interrupciones')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```
